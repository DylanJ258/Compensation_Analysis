```python
import pandas as pd

compensation = pd.read_csv('Compensation.csv')
print(compensation.head())
```

```python
columns = ['Department','Job', 'Employee Identifier', 'Total Salary','Retirement','Health and Dental','Other Benefits','Total Benefits', 'Total Compensation', 'Year']
compensation = compensation[columns]
compensation
```

```python
compensation = compensation[(compensation["Year"] == 2018) | (compensation["Year"] == 2019)| (compensation["Year"] == 2020)| (compensation["Year"] == 2021)| (compensation["Year"] == 2022)]
compensation
```

```python
compensation = compensation[(compensation["Department"] == 'Department of Technology')]
compensation
```

```python
compensation.to_csv('Technology_Compensation.csv')
```
